##  For the suburb Altona, it is postulated that a typical property sells for $800,000.

### Use the data at hand to test this assumption. Is the typical property price really $800,000 or has it increased? Use a significance level of 5%. 

```python
import pandas as pd
from scipy.stats import ttest_1samp

# Load the dataset
df = pd.read_excel(r"C:\Users\LENOVO\Downloads\property.xlsx")

df.head(10)
```

```python


# Filter for suburb Altona and remove null prices
altona_prices = df.loc[df['Suburb'] == 'Altona', 'Price'].dropna()

# Perform one-sample t-test
null_hypothesis_price =  500000
t_stat, p_value = ttest_1samp(altona_prices, null_hypothesis_price)

# Conclusion
if p_value / 2 < 0.05 and t_stat > 0:  # One-tailed test
    print(f"Reject the null hypothesis. The typical price is higher than $500,000.")
else:
    print(f"Fail to reject the null hypothesis. The typical price is not significantly higher than $500,000.")
```

```python
from scipy.stats import ttest_ind

# Filter data for 2016
df['Date'] = pd.to_datetime(df['Date'])
df_2016 = df[df['Date'].dt.year == 2016]

# Separate summer (April to September) and winter (October to March) prices
summer_prices = df_2016[df_2016['Date'].dt.month.isin([4, 5, 6, 7, 8, 9])]['Price'].dropna()
winter_prices = df_2016[df_2016['Date'].dt.month.isin([10, 11, 12, 1, 2, 3])]['Price'].dropna()

# Perform two-sample t-test
t_stat, p_value = ttest_ind(summer_prices, winter_prices, equal_var=False)

# Conclusion
if p_value < 0.05:
    print("There is a significant difference in property prices between summer and winter.")
else:
    print("There is no significant difference in property prices between summer and winter.")
```

```python
from scipy.stats import binom

# Calculate the probability of no car parking in Abbotsford
abbotsford = df[df['Suburb'] == 'Abbotsford']
p_no_parking = (abbotsford['Car'].isna().sum()) / len(abbotsford)

# Binomial probability
n, k = 10, 3
probability = binom.pmf(k, n, p_no_parking)

print(f"Probability of 3 out of 10 properties having no parking: {probability:.3f}")
```

```python
# Filter Abbotsford properties
prob_3_rooms = len(abbotsford[abbotsford['Rooms'] == 3]) / len(abbotsford)

print(f"Probability of finding a property with 3 rooms in Abbotsford: {prob_3_rooms:.3f}")
```
